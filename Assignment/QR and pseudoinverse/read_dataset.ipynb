```python
import yfinance as yf
from google.colab import drive
```

```python
# tickers was selected from 'List of largest companies by revenue'
# https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue
df = yf.download(tickers="WMT RDS-B AMZN F BP TM AAPL XOM CVS BRK-B UNH MCK BAC JPM DDAIF MSFT GOOG T ABC TOT",
                 period="16y",
                 interval="1wk")
```

```python
df.info()
```

```python
# We drop all rows that contains NaN values
df = df.dropna(axis=0)
df.info()
```

```python
df = df[['Adj Close', 'Volume']]
```

```python
df.columns.values
```

```python
# merge multiindex
df.columns = [' '.join(col) for col in df.columns.values]
df.columns.values
```

```python
df = df.reindex(columns=sorted(df.columns, key=lambda x: x.split()[-1]))
```

```python
# make 'date' column instead of index
df = df.reset_index()
```
