```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.linear_model import LinearRegression
from google.colab import drive
```

```python
drive.mount('/drive', force_remount=True)
df = pd.read_csv('/drive/My Drive/Colab Notebooks/York/numeric_analysis/Assignment/M/finances.csv')
df.head(5)
```

```python
df_aapl = df[['Date', 'Adj Close AAPL', 'Volume AAPL']]
df_aapl.head(5)
```

```python
df_aapl['Volume AAPL'] = df_aapl['Volume AAPL'].shift(1)
df_aapl.head(5)
```

```python
# convert `date` column to numerical values
df_aapl['Date'] = pd.to_datetime(df_aapl['Date'])
df_aapl['Date'] = df_aapl['Date'].map(dt.datetime.toordinal)
df_aapl.head(5)
```

```python
df_aapl = df_aapl.dropna()
```

```python
X = df_aapl[['Date', 'Volume AAPL']]
Y = df_aapl['Adj Close AAPL']

lm = LinearRegression()
model = lm.fit(X, Y)

print("score:", model.score(X, Y))
print("coefficients:", model.coef_)
print("intercept:", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error

preds = lm.predict(X)
mean_squared_error(Y, preds)
```
