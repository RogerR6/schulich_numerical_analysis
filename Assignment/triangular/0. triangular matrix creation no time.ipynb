## However triangular operations are supported in scipy.linalg


https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.solve_triangular.html

scipy.linalg.solve_triangular - Solve the equation ax = b for x, assuming a is a triangular matrix.
Parameters
a - triangular matrix
b - Right-hand side matrix in ax = b

lower - Use only data contained in the lower triangle of a. Default is to use upper triangle.

trans{0, 1, 2, 'N', 'T', 'C'}

0 or ‘N’ - ax = b

1 or ‘T’ - a^T x = b

2 or ‘C’ - a^H x = b
	

```
from scipy.linalg import solve_triangular
a = np.array([[3, 0, 0, 0], [2, 1, 0, 0], [1, 0, 1, 0], [1, 1, 1, 1]])
b = np.array([4, 2, 4, 2])
```

```
a
```

```
b
```

```
x = solve_triangular(a, b, lower=True)
x
```

```
a.dot(x)  # Check the result
```

```
A = np.array([
    [1, 1],
    [0, 1]
])
```

```
Ainv = solve_triangular(A, np.identity(A.shape[0]), lower=False)
Ainv
```

```
np.linalg.inv(A)
```
