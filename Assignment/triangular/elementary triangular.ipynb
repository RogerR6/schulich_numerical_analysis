```
# cycle
norms = []
for k in range(3, 100):
  sample = np.random.normal(loc = 0, scale = 1, size=(k, k))
  
  """
  variants of triangular matrix
  a = np.triu(sample) +
  a = np.tril(sample) +
  a = np.triu(sample, 1) - becouse singular matrix 
  a = np.triu(sample, -1) - not equal
  """
  a = np.triu(sample)
  b = np.random.normal(loc = 0, scale=1, size=(k))
  # print("a = ")
  # print(a)
  # print("---------------------------------------------------")
  # print("b = ")
  # print(b)
  # print("---------------------------------------------------")
  xSimple = np.linalg.solve(a, b)
  xTriangular = solve_triangular(a, b)
  # xTriangular = solve_triangular(a, b, lower=True)
  # print("Result of solving is - {}".format(np.array_equal(xSimple, xTriangular)))
  # if np.array_equal(xSimple, xTriangular):

  # print("Product is - ")
  # print(a.dot(xSimple))

  # print("xSimple = ")
  # print(xSimple)
  # print("xTriangular = ")
  # print(xTriangular)
  # print("")
  # print("Delta -")
  deltaX = xSimple - xTriangular
  # print(deltaX)
  normDelta = np.linalg.norm(deltaX)
  sub = []
  sub.append(k)
  sub.append(normDelta)
  norms.append(sub)
  print("iteration {} - normDelta={}  - ".format(k, normDelta))
  # print("=========================================================================")
```

```
normsD = []

normsSimple = []
normsTrinangular = []

for k in range(3, 100):
  sample = np.random.normal(loc = 0, scale = 1, size=(k, k))
  
  """
  variants of triangular matrix
  a = np.triu(sample) +
  a = np.tril(sample) +
  a = np.triu(sample, 1) - becouse singular matrix 
  a = np.triu(sample, -1) - not equal
  """
  a = np.triu(sample)
  # b = np.random.normal(loc = 0, scale=1, size=(k))
  xTrue = np.ones(k)

  b = a @ xTrue

  

  xSimple = np.linalg.solve(a, b)
  xTriangular = solve_triangular(a, b)

  
  deltaX = xSimple - xTriangular
  errorSimple = xSimple - xTrue
  errorTriangular = xTriangular - xTrue


  # print(deltaX)
  normDelta = np.linalg.norm(deltaX)
  normSimple = np.linalg.norm(errorSimple)
  normTriangular = np.linalg.norm(errorTriangular)


  sub = []
  sub.append(k)
  sub.append(normDelta)
  normsD.append(sub)

  sub = []
  sub.append(k)
  sub.append(normSimple)
  normsSimple.append(sub)

  sub = []
  sub.append(k)
  sub.append(normTriangular)
  normsTrinangular.append(sub)


  # print("iteration {} - normDelta={}  - ".format(k, normDelta))
```

```
pyplot.subplot(1, 1, 1)
# Plotting the graph
pyplot.plot(normsD, color='blue', label="DeltaNorm")
pyplot.plot(normsSimple, color='red', label="normsSimple")
pyplot.plot(normsTrinangular, color='green', label="normsTrinangular")
# Setting a logarithmic scale for y-axis
pyplot.legend()
# pyplot.yscale('log')
pyplot.show()
```
