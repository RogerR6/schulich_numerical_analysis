```
import pandas as pd
import requests
```

```
# declare data sources
facebook_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRkCEf20V8iCyjql6mEHef16dhS6-ZShrxnUtauxqa6MnKeOZ7O6LyTMCDnkUrknWRIViPLp6xSha6o/pub?output=csv"
microsoft_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vTKMlwwlAc8o20GW3uoE-EftfmpKy1ngsqPnAIGw_-O9EkFgY8XHcMPLwKXg0bp0kKEErLMGyp-inLC/pub?gid=0&single=true&output=csv"
tesla_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRRZpclS-nzUhfcWrvCBnvqFTrGBmaCoxRpwO5vsFNgqxWNHtEP8f4Uia9hYWN8G0S5IHeEgkg-NAxZ/pub?output=csv"
```

```
df_facebook = pd.read_csv(facebook_path, encoding="utf8")
df_microsoft = pd.read_csv(microsoft_path, encoding="utf8")
df_tesla = pd.read_csv(tesla_path, encoding="utf8")
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plot
```

```
#prepare x and y
x_facebook = df_facebook[["Open",	"High",	"Low", "Close"]].values
y_facebook = df_facebook["Volume"].values
print(len(x_facebook), len(y_facebook))

x_microsoft = df_microsoft[["Open",	"High",	"Low", "Close"]].values
y_microsoft = df_microsoft["Volume"].values

x_tesla = df_tesla[["Open",	"High",	"Low", "Close"]].values
y_tesla = df_tesla["Volume"].values
```

```
df_facebook
```

```
# declare train test split
xTrain_facebook, xTest_facebook, yTrain_facebook, yTest_facebook = train_test_split(x_facebook, y_facebook, test_size = 0.2, random_state = 0)
xTrain_microsoft, xTest_microsoft, yTrain_microsoft, yTest_microsoft = train_test_split(x_microsoft, y_microsoft, test_size = 0.2, random_state = 0)
xTrain_tesla, xTest_tesla, yTrain_tesla, yTest_tesla = train_test_split(x_tesla, y_tesla, test_size = 0.2, random_state = 0)
```

```
# declare regressions
"""
1 - facebook
2 - microsoft
3 - tesla

"""
linearRegressor1 = LinearRegression()
linearRegressor2 = LinearRegression()
linearRegressor3 = LinearRegression()
```

```
# fitting
linearRegressor1.fit(xTrain_facebook, yTrain_facebook)
linearRegressor2.fit(xTrain_microsoft, yTrain_microsoft)
linearRegressor3.fit(xTrain_tesla, yTrain_tesla)
```

```
# predicts originals
yPrediction_facebook = linearRegressor1.predict(xTest_facebook)
yPrediction_microsoft = linearRegressor2.predict(xTest_microsoft)
yPrediction_tesla = linearRegressor3.predict(xTest_tesla)
```
