```python
def f1(f):
    print("f1 received f =", f)
    result = f()
    print("Result of f invocation is", result)
    
```

```python
f = lambda: 5
f1(f)
```

Now with arguments

```python
def f2(f, a):
    print("f2 received f =", f)
    print("Its single argument a =", a)
    result = f(a)
    print("The result of f invocation with argument a is", result)
```

```python
f3 = lambda a: a**2
f2(f3, 5)
```

## Args

I have a function f3 with a single argument a. 
Instead of giving it by name, I want to give it as an array. 


```python
f3((5)), f3(((5)))
# Parentheses are dropped
```

Now 2 arguments

```python
f4 = lambda a, b: a+b
```

```python
f4(5, 6)
```

```python
# Now the list
argList = [5, 6]
f4(*argList)
```

kwargs

```python
argDict = {"a": 5, "b": 6}
f4(*argDict), f4(**argDict)
```

```python
def f5(**kwargs):
    print("kwargs['a'] =", kwargs["a"])
    print("kwargs['b'] =", kwargs["b"])
```

```python
f5(a = 5, b = 6)
```

The opposite case

```python
def f6(a = 1, b = 2):
    print ("I am f6, and I am printing a+b =", a + b)
```

```python
f6(**argDict)
```

Args again. It is positional

```python
f6(*argList)
```
