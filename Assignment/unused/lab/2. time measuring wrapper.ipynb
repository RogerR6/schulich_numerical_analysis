Length of some operation

```python
startSeconds = time()
2**100
endSeconds = time() 
duration = endSeconds - startSeconds
duration
```

```python
def computeResultAndDuration():
    startSeconds = time()
    result = 2**100
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
result, duration = computeResultAndDuration()
print("result =", result)
print("duration =", duration)
```

```python
def computeResultAndDurationOfFunction(f):
    startSeconds = time()
    result = f()
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
def computationF():
    result = 10**100 + 10**100 + 10**100  
    return result
    
```

```python
result, duration = computeResultAndDurationOfFunction(computationF)
result, duration
```

```python
def computeResultAndDurationOfArgFunction(f, n):
    startSeconds = time()
    result = f(n)
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
def computationArgF(n):
    s = 0
    for i in range(n):
        s += 1  
    return s
```

```python
result, duration = computeResultAndDurationOfArgFunction(computationArgF, 10000000)
result, duration
```

```python
def resDurArgs(f, *args):
    startSeconds = time()
    result = f(*args)
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
result, duration = resDurArgs(computationArgF, 10000000)
result, duration
```

```python
def resDurKwargs(f, **kwargs):
    startSeconds = time()
    result = f(**kwargs)
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
result, duration = resDurKwargs(computationArgF, n=10000000)
result, duration
```

```python
def resDurArgsKwargs(f, *args, **kwargs):
    startSeconds = time()
    result = f(*args, **kwargs)
    endSeconds = time() 
    duration = endSeconds - startSeconds
    return result, duration
```

```python
result, duration = resDurArgsKwargs(computationArgF, n=10000000)
result, duration = resDurArgsKwargs(computationArgF, 10000000)
result, duration
```
