```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1ozmJBOaE2Xh6gjDGqy4TI27XIhyAL6h60QRy1MUsDso/pub?output=csv')
stringDateSeries = df["Date"]
stringDateSeries.head(5)
```

```python
type(stringDateSeries), type(stringDateSeries[0]) 
```

```python
firstDate = stringDateSeries[0]
firstDate
```

```python
# convert `date` column to numerical values
# more precisely number of days since 1st january 1AD
dateSeries = pd.to_datetime(stringDateSeries)
dateSeries.head(5)
```

```python
type(dateSeries), type(dateSeries[0]), 
```

```python
ordinalDateSeries = dateSeries.map(dt.datetime.toordinal)
ordinalDateSeries.head(5)
```

```python
type(ordinalDateSeries), type(ordinalDateSeries[0])
```

```python
weeklyIncrement1 = ordinalDateSeries[1] - ordinalDateSeries[0]
weeklyIncrement1
```

```python
weeklyIncrement2 = ordinalDateSeries[2] - ordinalDateSeries[1]
weeklyIncrement2
# So ordinal is the number or days since 1AD
```

Now let us work on map, the way to apply the same function to each object in a collection of objects. 

Let us experiment in the following scenario. Create a list of 2 numbers. And we need to increment those numbers. 

```python
myList = [1, 2]
```

```python
def increment(a):
    return a+1
```

```python
myIterable = map(increment, myList)
list(myIterable)
```
