```python
import matplotlib.pyplot as plt
import numpy as np
```

In matplotlib we have several methods to draw pictures (multi-panel plots):

    fig.add_axes() - base method, convenient for nested diagrams;

    fig.add_subplot() - adds a plot on the canvas. convenient to show 2-3 diagrams;

    plt.subplot() - similar to the previous one, but for pyplot;

    plt.subplots() - convenient method for coreation of several subplots;

    plt.GridSpec() - method to unite cells insubplots for more complicated configurations. Allows to create subplots of the different form on the picture.


```python
# simple example
fig = plt.figure()
# In 111, first one - number of graphs,
# 2nd one is a row, and third is a column. 
ax = fig.add_subplot(111) 
# ax = fig.add_subplot(112) - will not work
fig.set(facecolor = 'green')
ax.set(facecolor = 'red')

plt.show()
```

```python
# Same as in the previous example, only with add_axies
fig = plt.figure()
# Takes absolute?
# ignores some numbers?
box = [0.5, 0.15, 0.78, 0.78] 
# This parameter is the dimensions [left, bottom, width, height] of the new axes.
box = [0.1, 0.1, 0.3, 0.3]
# Creation of a sample Axes c by Figure-method add_axes()
ax = fig.add_axes(box)
fig.set(facecolor = 'green')
ax.set(facecolor = 'red')
```

```python
# Creation of a figure
fig = plt.figure()
# Adding a subplot
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx() # Creation of a common axis
t = np.linspace(0., 10., 100)
ax1.plot(t, t ** 2, 'b-')
ax2.plot(t, 1000 / (t + 1), 'r-')
ax1.set_ylabel('Density (cgs)', color='red')
ax2.set_ylabel('Temperature (K)', color='blue')
ax1.set_xlabel('Time (s)')
ax2.set_xlabel('something else') # went nowhere
plt.show()
```

```python
# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

```python
fig = plt.figure()

# Creation of instance of Axes by Figure-method add_subplot()
ax = fig.add_subplot(111)
# Or this way
box = [0.25, 0.5, 0.25, 0.25]
# Creation of instance of Axes by Figure-method add_axes()
ax2 = fig.add_axes(box)

x = np.arange(0.0, 1.0, 0.1)
y = np.sin(x)*np.exp(x)
z = np.cos(x)*np.sin(x)

# Method plot() gets invoked from an instance of ax, not plt (interface pyplot)
ax.plot(y)
ax.plot(z)

ax2.plot(y)

for ax in fig.axes:
    ax.grid(True)
```
