```
import math

# issue with using == operator in floating point arithmetics
# use math.isclose instead
a = 1/3.0
b = 0.1/0.3
print("a= 1/3.0", a, ",\nb= 0.1/0.3", b, ",\nCompare a==b: ", a == b)
print("Using isclose:", math.isclose(a, b))

print("------------------")
orig = 10.0
a = orig / 77
back = a * 77
print("orig= ", orig, ",\nback= orig / 77*77= ", back, ",\nCompare orig==back: ", orig == back)
print("Using isclose:", math.isclose(orig, back))
```

```
# adding small values to big cancels small
print("\nSummation example")
small = 0.00001
big = 10 ** 20
neg_big = -big

print("small+big+neg_big= ", small+big+neg_big)
print("big+neg_big+small= ", big+neg_big+small)
print("big+small+neg_big= ", big+small+neg_big)
print("neg_big+small+big= ", neg_big+small+big)
```

```
def addToOneError(count):
    term = 1/count
    sum = 0.
    for _ in range(count):
        sum += term
    error = sum - 1
    return error        
```
