```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

x = sympy.Symbol("x", real=True)
F = sympy.Lambda(x, sympy.sin(x) / x)

xpts = np.linspace(0, 50, 1000)

ypts = list(map(F, xpts))
plt.plot(xpts, ypts)
```
