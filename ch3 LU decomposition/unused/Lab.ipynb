```python
import scipy.linalg
A = scipy.array([[1, 2, 3],
             [4, 5, 6],
             [10, 11, 9]])
P, L, U = scipy.linalg.lu(A)
print(P)
print(L)
print(U)
```

```python
import numpy as np
A = np.array([[1, 2, 3],
             [4, 5, 6],
             [10, 11, 9]])
P, L, U = scipy.linalg.lu(A)
mult = P.dot((L.dot(U)))
print(mult)
```
