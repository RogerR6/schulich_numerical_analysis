LU with scipy

scipy.linalg.lu returns 

$P$ permutation matrix,

$L$ lower-unitriangular matrix,

$U$ upper-triangular matrix

Applies partial pivoting to any input matrix

```
from scipy import linalg as la
import numpy as np
```

```
A = np.array([[2, 3], [5, 4]])
b = np.array([4, 3])
P, L, U = la.lu(A)
L
```

```
U
```

```
P
```

```
P.dot(L.dot(U))
```

```
la.solve(A, b)
```

```
import sympy
sympy.init_printing()
```

```
# try a matrix which requires permutation
# when reducing A to upper-triangular zerows appear 2 times on the diagonal
# hence 2 permutations are required
# [0,1] swaps the 1-st and 2-nd rows at the first step of elimenation
# [1,2] swaps the 2-nd and 3-d rows at the second step

A = sympy.Matrix([[0, 0, 2], [1, 3, 2], [1, 2, 3]])
L, U, P = A.LUdecomposition()
P
```

```
# works for degenerate matrix
A = sympy.Matrix([[0, 0,], [1, 3]])
L, U, P = A.LUdecomposition()
print('L:\n', L)
print('U:\n', U)
print('P:\n', P)
```

```
# both wide and tall matrices also work
A = sympy.Matrix([[1, 2], [4, 5], [7, 9]])
L, U, P = A.LUdecomposition()
print('L:\n', L)
print('U:\n', U)
print('P:\n', P)
```

```
# problems with small pivot
eps = 10**(-18)
A = sympy.Matrix([[eps, 1], [1, 1]])
L, U, P = A.LUdecomposition()
print('L=\n', L)
print('U=\n', U)
print('P=\n', P)
LU = L@U
print('A=\n', A)
print('LU=\n', LU)
# the product LU is not equal to the original matrix A
```
