```python
# practice
A3 = np.array([[1, 1], [1, 2], [1, 3]])
# mt
# A3 = np.array([[1, 1], [1, 0], [1, -1]])
A3
```

```python
# practice
b3 = np.array([[3], [5], [1]])
# mt
# b3 = np.array([[2], [3], [4]])
b3
```

```python
reg = LinearRegression(fit_intercept=False)
reg.fit(X=A3, y=b3)
reg.coef_
```

```python
AtA = A3.T @ A3
AtA
```

```python
det(AtA)
```

```python
AtAinv = np.linalg.inv(AtA)
AtAinv
```

```python
AtB = A3.T @ b3
AtB
```

```python
solve(AtA, AtB)
```

```python
AtAinv @ AtB
```

Tikhonov sustem formed by hand

```python
AtATikhonov = AtA + 1 * np.eye(AtA.shape[0])
# np.array([[4, 6], [6, 15]])
AtATikhonov
```

```python
solve(AtATikhonov, AtB)
```
