```python
import numpy as np
from scipy import linalg
```

```python
A = np.array([[2, 6], 
              [-1, 2]])
b = np.array([[4], 
              [3]])
L = np.array([[2, 0], 
              [-1, 5]])
U = np.array([[1, 3], 
              [0, 1]])

```

```python
L @ U
```

```python
y = linalg.solve(L, b)
y
```

```python
linalg.solve(U, y)
```

```python
linalg.solve(A, b)
```

# Linear regression on three observations and Tikhonov regularization

```python
X = np.array([[1, 1], [1, 0], [1, -1]])
y = np.array([[2], [3], [-2]])
XT = X.T
XTX = XT @ X
XTX
```

```python
XTy = XT @ y
XTy
```

```python
np.linalg.det(XTX)
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
x_ls = XTX_inv @ XTy
x_ls
```

```python
np.linalg.norm(x_ls)
```

```python
alpha = 1
X_reg = XTX + alpha * np.identity(2)
X_reg
```

```python
x_reg = linalg.solve(X_reg, XTy)
x_reg
```

```python
np.linalg.norm(x_reg)
```
