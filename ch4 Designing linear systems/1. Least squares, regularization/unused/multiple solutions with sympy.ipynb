```python
import sympy as sym
import numpy as np

x,y, z = sym.symbols('x,y,z')
eq1 = sym.Eq(x+3*y-z,4)
eq2 = sym.Eq(0*x-13*y+6*z,-8)
eq3= sym.Eq(0*x+0*y+0*z,0)
eq1
```

```python
eq2
```

```python
eq3
```

```python
result = sym.solve([eq1,eq2,eq3],(x,y,z))
print(result)
```
