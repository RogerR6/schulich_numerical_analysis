#Cholesky 2 by 2 other numbers

```
A = np.array([[9, 3], [3, 17]])
L = scipy.linalg.cholesky(A, lower=True)
L
```

```
U = scipy.linalg.cholesky(A, lower=False)
U
```

```
L @ L.T
```

```
E1 = np.array([[1/3, 0], [-1/3, 1]])
E1 @ A
```

```
E1 @ A @ E1.T
```

```
E2 = np.array([[1, 0], [0, 1/4]])
E2 @ E1 @ A @ E1.T
```

```
E2 @ E1 @ A @ E1.T @ E2.T
```

```
L = np.linalg.inv(E1) @ np.linalg.inv(E2)
L
```

```
L @ L.T
```
