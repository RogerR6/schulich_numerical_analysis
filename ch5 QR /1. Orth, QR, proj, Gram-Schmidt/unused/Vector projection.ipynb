```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
from skspatial.objects import Plane
from skspatial.objects import Vector
from skspatial.plotting import plot_3d
```

```python
plane = Plane(point=[0, 0, 0], normal=[0, 0, 1])
vector = Vector([1, 1, 1])

vector_projected = plane.project_vector(vector)


_, ax = plot_3d(
    plane.plotter(lims_x=(-5, 5), lims_y=(-5, 5), alpha=0.1),
    vector.plotter(point=plane.point, color='k'),
    vector_projected.plotter(point=plane.point, color='r', linewidth=2, zorder=3),
)

ax.set_zlim([-1, 1])
```

```python
_, ax = plot_3d(
    vector.plotter(point=plane.point, color='k'),
    vector_projected.plotter(point=plane.point, color='r', linewidth=2, zorder=3),
)
```

```python
vector = Vector([1, 0, 1])
vector_projected = plane.project_vector(vector)

_, ax = plot_3d(
    vector.plotter(point=plane.point, color='k'),
    vector_projected.plotter(point=plane.point, color='r', linewidth=2, zorder=3),
)
```

```python
from skspatial.objects import Line

line = Line((1, 0, 0), (1, 1, 0))
vector = Vector([1, 2, 3])

vector_projected = line.project_vector(vector)


plot_3d(
    line.plotter(t_1=-1, c='k', linestyle='--'),
    vector.plotter(point=line.point, color='k'),
    vector_projected.plotter(point=line.point, color='r', linewidth=2, zorder=3),
)
```

connect the third side of the triangle

```python
line = Line((1, 0, 0), (1, 1, 0))
vector = Vector([1, 2, 3])

vector_projected = line.project_vector(vector)
side3beginning = vector # no
side3end = vector_projected # no
side3vector = Vector.from_points(side3beginning, side3end) # no

plot_3d(
    line.plotter(t_1=-1, c='k', linestyle='--'),
    vector.plotter(point=line.point, color='k'),
    vector_projected.plotter(point=line.point, color='r', linewidth=2, zorder=3),
    side3vector.plotter(point=line.point, color='g', linewidth=2, zorder=3), # no
)
```

```python
from skspatial.objects import Plane
from skspatial.objects import Point
from skspatial.objects import Vector
from skspatial.plotting import plot_3d



point = Point([5, 9, 3])

point_projected = plane.project_point(point)
vector_projection = Vector.from_points(point, point_projected)


plot_3d(

    point.plotter(s=75, c='k'),
    point_projected.plotter(c='r', s=75, zorder=3),
    vector_projection.plotter(point=point, c='k', linestyle='--'),
)
```

```python
from skspatial.objects import Plane
from skspatial.objects import Vector
from skspatial.plotting import plot_3d


vector = Vector([1, 1, 1])

vector_projected = plane.project_vector(vector)
point = Point(vector)

point_projected = plane.project_point(vector)
vector_residual = Vector.from_points(point, point_projected)

_, ax = plot_3d(
    vector.plotter(point=plane.point, color='k'),
    point_projected.plotter(c='r', s=75, zorder=3),
    point.plotter(s=75, c='k'),
    vector_projected.plotter(point=plane.point, color='r', linewidth=2, zorder=3),
    vector_residual.plotter(point=point, c='k', linestyle='--'),

)

ax.set_zlim([-1, 1])
```
