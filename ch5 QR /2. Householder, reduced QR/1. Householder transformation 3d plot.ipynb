```python
from skspatial.objects import Plane
from skspatial.objects import Vector
from skspatial.plotting import plot_3d
from skspatial.objects import Line
from skspatial.objects import Point


from google.colab import output
output.enable_custom_widget_manager()

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
%matplotlib widget
```

```python
plane = Plane([0, 0, 0], [0, 0, 1])
vector1 = Vector([1, 1, 1])
vector2 = Vector([1, 1, -1])
vector3 = Vector([-1, -1, 1])
line = Line(point=[0, 0, 0], direction=[0, 0, 1])
vector4 = Line(vector1, vector2 - vector1)

point = Point([0, 0, 0])

vector_projected = plane.project_vector(vector1)


_, ax = plot_3d(
    plane.plotter(lims_x=(-5, 5), lims_y=(-5, 5), alpha=0.3),
    vector1.plotter(point=plane.point, color='k'),
    vector2.plotter(point=plane.point, color='b'),
    vector3.plotter(point=plane.point, color='g', linestyle='--'),
    line.plotter(t_2=5),
    vector4.plotter(color='y', linestyle='--'),
    vector_projected.plotter(point=plane.point, color='r', linewidth=2, zorder=3),
    point.plotter(s=75, c='k'),
)

ax.set_zlim([-1, 1])
plt.show()
```
