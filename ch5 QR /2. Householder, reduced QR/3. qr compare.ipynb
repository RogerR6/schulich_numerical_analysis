Numpy QR

https://numpy.org/doc/stable/reference/generated/numpy.linalg.qr.html


```python
A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
A
```

In numpy the option mode{‘reduced’, ‘complete’, ‘r’, ‘raw’}, optional

In scipy the same option mode{‘full’, ‘r’, ‘economic’, ‘raw’},

Both have the first values as default. 

In case of tall matrix A, m by n, m>=n, scipy makes the suare matrix Q of the bigger side size m. 

Numpy, by default, reduces the Q matrix. And does some correspondent changes to R. 

The default of sympy is kind of scipy's full or numpy's complete. As to option specification, I did not find any. 


```python
q, r = np.linalg.qr(A)
q
```

```python
np.linalg.qr(A)
```

```python
r
```

```python
q, r = np.linalg.qr(A, mode="raw")
q
```

```python
r
```

Shapes in the rectangular cases. Full and reduced

```python
import numpy as np
A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
q, r = np.linalg.qr(A)
q, np.shape(q)
```

```python
r, np.shape(r)
```

```python
import numpy as np
import scipy

A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
q, r = scipy.linalg.qr(A)
q, np.shape(q)
```

```python
r, np.shape(r)
```

```python
import numpy as np
A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
q, r = np.linalg.qr(A, mode="complete")
q, np.shape(q)
```

```python
r, np.shape(r)
```

```python
import numpy as np
import scipy

A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
q, r = scipy.linalg.qr(A, mode="economic")
q, np.shape(q)
```

```python
r, np.shape(r)
```

```python
import numpy as np
import scipy

A = np.array([[0, 1], [1, 1], [1, 1], [2, 1]])
q, r = np.linalg.qr(A, mode="reduced")
q, np.shape(q)
```
