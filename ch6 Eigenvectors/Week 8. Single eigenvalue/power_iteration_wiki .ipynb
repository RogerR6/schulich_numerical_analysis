```
import numpy as np
float_formatter = lambda x: "%.2f" % x
np.set_printoptions(formatter={'float_kind':float_formatter})
def power_iteration(A, num_simulations: int):
    # Ideally choose a random vector
    # To decrease the chance that our vector
    # Is orthogonal to the eigenvector
    b_k = (np.random.rand(A.shape[1]) - 1/2)*10
    # b_k = np.random.normal(A.shape[1])

    for iteration in range(num_simulations):
        # calculate the matrix-by-vector product Ab
        b_k1 = np.dot(A, b_k)

        # calculate the norm
        b_k1_norm = np.linalg.norm(b_k1)

        # re normalize the vector
        b_k = b_k1 / b_k1_norm
        print("iteration: ", iteration, ", b_k = ", b_k)
        numerator = b_k1 @ A @ b_k
        denominator = b_k @ b_k
        lambda_k = numerator / denominator
        print("lambda_k = ", lambda_k)
        print()
    return b_k

power_iteration(np.array([[0.5, 0.5], [0.2, 0.8]]), 10)
```
