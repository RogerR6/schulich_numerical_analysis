```
import numpy as np
from numpy import linalg
```

```
# A = np.array([
#               [-6, 3],
#               [4, 5]
# ])

A = np.array([
              [1, -2,-1],
              [1, 4, 2],
              [0, -6, -4]
])
coeffs = np.poly(A)
coeffs
```

```
linalg.eig(A)
```

```
# given coefficients, find roots
roots = np.roots(coeffs)
roots
```

```
eigenvectorCandidate = np.array([1, 4, 1])
A @ eigenvectorCandidate
```

```
import numpy as np
a = np.array([[3, 1], [2, 2]])
w, v = np.linalg.eig(a)
np.isclose(np.dot(a, v), w*v)
```

Example 6.3 (QR iteration). To illustrate typical behavior of QR iteration, we apply the algorithm to the matrix

```
A1 = np.array([
               [2, 3],
               [3, 2]
])
x1Hat = np.array([0.555, 0.832])
x2Hat = np.array([0.832, -0.555])
A1 @ x1Hat
```

```
A1 @ x2Hat
```

```
np.linalg.eig(A1)
```
