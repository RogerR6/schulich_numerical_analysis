```
import numpy as np
a = np.array([[3, 1], [2, 2]])
w, v = np.linalg.eig(a)
print(w)
print(v)
```

How does dot differ from *?

```
np.dot(a, v) == w*v
```

```
w*v
```

```
np.dot(a, v)
```

```
np.array_equal(np.dot(a, v), (w*v))
```

```
# !!!!
np.allclose(np.dot(a, v), w*v)
```

```
np.isclose(np.dot(a, v), w*v)
```
