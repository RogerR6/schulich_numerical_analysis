```python
import numpy as np
from sympy.matrices import Matrix
```

```python
#A = np.array([[1,2], [2, 3], [5, 6]])

A = np.array([[0, 2], [0, 0], [1, 0]])
```

```python
u, s, vt = np.linalg.svd(A)
u_p = Matrix(np.round(u, 3))
s_p = Matrix(np.round(s, 3))
vt_p = Matrix(np.round(vt, 3))
```

```python
u_p
```

```python
s_p
```

```python
vt_p
```

```python
S = np.zeros((3, 2))
S[0, 0] = s.flatten()[0]
S[1, 1] = s.flatten()[1]
```

```python
Matrix(np.round(S, 3))
```

```python
Matrix(u @ S @ vt)
```

```python
AtA = A.T @ A
Matrix(AtA)
```

```python
egenvalues, egenvectors = np.linalg.eig(AtA)
Matrix(np.round(np.sqrt(egenvalues), 3))
```

```python
Matrix(np.round(egenvectors, 3))
```

```python
AAt = A @ A.T
Matrix(np.round(AAt, 3))
```

```python
egenvalues, egenvectors = np.linalg.eig(AAt)
Matrix(np.round(egenvalues, 3))
```

```python
Matrix(np.round(egenvectors, 3))
```

```python
# Check that eigenvectors are orthogonal
Matrix(np.round(egenvectors @ egenvectors.T, 3))
```

```python
U = np.array([[1, 0, 0],
              [0, 0, 1],
              [0, 1, 0]])

S = np.array([[2, 0],
              [0, 1],
              [0, 0]])

V = np.array([[0, 1],
              [1, 0]])
Matrix(U)
```

```python
Matrix(S)
```

```python
Matrix(V)
```

```python
A = U @ S @ V.T
Matrix(A)
```

```python
u, s, vt = np.linalg.svd(A)
```

```python
Matrix(u)
```

```python
Matrix(s)
```
