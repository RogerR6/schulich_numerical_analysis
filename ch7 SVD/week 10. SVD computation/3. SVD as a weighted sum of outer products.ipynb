3 matrices - any * diag * any

A * D * B

```python
A = np.array([
              [1, 2], 
              [3, 4]
])
B = np.array([
              [5, 6], 
              [7, 8]
])
D = np.diag([10, 20])
D
```

```python
ADBdirect = A @ D @ B
ADBdirect
```

```python
outer1 = np.outer(A[:, 0], B[0, :])
outer1
```

```python
outer2 = np.outer(A[:, 1], B[1, :])
outer2
```

```python
ADBbyOuterWeightedSum = D[0, 0] * outer1  + D[1, 1] * outer2
ADBbyOuterWeightedSum
```

```python
ADBbyOuterWeightedSum == ADBdirect
```

```python
def product_by_sum_of_outer(A, B):
    if A.shape[1] == B.shape[0]:
        rows1, cols1 = A.shape
        rows2, cols2 = B.shape
        outer_product = np.zeros((rows1, cols2))
        for i in range(0, cols1):
            # print(A[:, i].reshape(-1, 1) @ B[i, :].reshape(1, cols))
            outer_product += A[:, i].reshape(-1, 1) @ B[i, :].reshape(1, cols2)
            print("--------------")
        print(outer_product)
        return outer_product
    else:
        print("incompatible shapes")
        print(A.shape)
        print(B.shape)
        return False
```

```python
A = np.array([
              [1, 2], 
              [3, 4]
])
B = np.array([
              [5, 6], 
              [7, 8]
])

abBySumOfOuter = product_by_sum_of_outer(A,B)
```

```python
abDirectly = A @ B
abDirectly
```

```python
np.isclose(abBySumOfOuter, abDirectly)
```
