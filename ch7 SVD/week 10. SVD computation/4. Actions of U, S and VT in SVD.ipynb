```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
i = np.array([1, 0])
j = np.array([0, 1])
x = i + j
```

Expand A into SVD and check it by multiplication

```python
A = np.array([[2, 3], [-2, 4]])
U, s, V_t = np.linalg.svd(A)

S = np.diag(s)

US = np.dot(U, S)
USV_t = np.dot(US, V_t)
USV_t
```

Below V is a standard basis matrix or identity matrix. It is not V matrix of the SVD. 

```python
I = np.array([i, j])
I
```

```python
UI = np.dot(U, I)
UI
```

```python


plt.arrow(0, 0, UI[0][0], UI[0][1], color='b', label='Ui', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, UI[1][0], UI[1][1], color='b', label='Uj', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, i[0], i[1], color='r', label='i', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, j[0], j[1], color='r', label='j', head_length=.2, head_width=.1, length_includes_head=True)

plt.title('First rotation in SVD')
plt.xlim(-3, 3)
plt.ylim(-2, 2)


plt.grid()

plt.rcParams["figure.figsize"] = (15,10)
```

In the above picture, standard basis i and j is shown in red. The blue vectors are columns of U matrix. It is the same as actions of U operator on standard basis i and j vectors. 


In the picture below, we show action of S matrix onto the standard basis. 

```python
I = np.array([i, j])
SI = np.dot(S, I)

plt.arrow(0, 0, SI[0][0], SI[0][1], color='b', label='Si', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, SI[1][0], SI[1][1], color='b', label='Sj', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, i[0], i[1], color='r', label='i', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, j[0], j[1], color='r', label='j', head_length=.2, head_width=.1, length_includes_head=True)

plt.title('Stretches in SVD')
plt.xlim(-1, 6)
plt.ylim(-1, 4)
plt.legend()

plt.rcParams["figure.figsize"] = (10,5)
plt.grid()
```

```python
I = np.array([i, j])
VTI = np.dot(V_t, I)

plt.arrow(0, 0, VTI[0][0], VTI[0][1], color='b', label='VTi', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, VTI[1][0], VTI[1][1], color='b', label='VTj', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, i[0], i[1], color='r', label='i', head_length=.2, head_width=.1, length_includes_head=True)
plt.arrow(0, 0, j[0], j[1], color='r', label='j', head_length=.2, head_width=.1, length_includes_head=True)

plt.title('Second rotation in SVD')
plt.xlim(-1, 6)
plt.ylim(-1, 4)
plt.legend()

plt.rcParams["figure.figsize"] = (10,5)
plt.grid()
```
