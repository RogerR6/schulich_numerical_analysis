```python
import numpy as np
from sklearn.decomposition import PCA
from numpy.linalg import svd

```

```python
X = np.array([
              [1, 2, 3], 
              [5, 2, -4], 
])
x_row_means = X.mean(axis=1).reshape([2, 1])
x_row_means
```

```python
X_centered = X - x_row_means
X_centered
```

```python
U, s, VT = svd(X)
U
```

```python
Z_by_svd = U.T @ X
Z_by_svd
```

```python
U_centered, s_centered, VT_centered = svd(X_centered)
U_centered
```

```python
Z_centered_by_svd = U_centered.T @ X_centered
Z_centered_by_svd
```

```python
pca = PCA()
Z_by_pca_T = pca.fit_transform(X.T)
Z_by_pca = Z_by_pca_T.T
Z_by_pca
```

```python
import numpy as np
from sklearn.decomposition import PCA
from numpy.linalg import svd

X = np.array([
              [-1, -4, -1, -2], 
              [1, 0, -1, 4], 
])
x_row_means = X.mean(axis=1).reshape([2, 1])
X_centered = X - x_row_means
U_centered, s_centered, VT_centered = svd(X_centered)
Z_centered_by_svd = U_centered.T @ X_centered
Z_centered_by_svd
print(Z_centered_by_svd)
pca = PCA()
Z_by_pca_T = pca.fit_transform(X.T)
Z_by_pca = Z_by_pca_T.T
print(Z_by_pca)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from numpy.linalg import svd
X = np.array([
              [-1, -4, -1, -2], 
              [1, 0, -1, 4], 
])
U, s, VT = svd(X)
Z_by_svd = U.T @ X
print(Z_by_svd)
pca = PCA()
Z_by_pca_T = pca.fit_transform(X.T)
Z_by_pca = Z_by_pca_T.T
print(Z_by_pca)
```
