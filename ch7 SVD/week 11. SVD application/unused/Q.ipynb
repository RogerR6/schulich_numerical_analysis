```python
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1, -1], 
              [-2, -1], 
              [-3, -2], 
              [1, 1], 
              [2, 1], 
              [3, 2]])
X = X + 10
pca = PCA(n_components=2)
pca = PCA()
pca.fit(X)

# eigenvectors
print(pca.components_)
print()
# eigenvalues
print(pca.explained_variance_)
print()
print(np.sqrt(pca.explained_variance_))
```

```python
# https://aaronschlegel.me/matrix-norms-inequalities-python.html

import numpy as np
A = np.random.randint(-10, 10, size=(3, 2))
n, m = A.shape

f = 0

for i in np.arange(n):
    for j in np.arange(m):
        f = f + np.sum(np.power(np.abs(A[i, j]), 2))

np.sqrt(f)
```

```python
# https://medium.com/@avecillasprince/power-method-with-python-15b0e85ee5c9
import numpy as np

A = np.array([[0.9,10.2,0.7],
                [3.3,0.9,1.2],
                [0.01,7.4,2.5]],float)

u = np.array([[1],[1],[1]], float)

n = 100
eigenvalue = 0

for i in range(n):
    u = A @ u
    eigenvalue = np.max(u)
    u = u / eigenvalue

print('eigenvector: \n', u)
print('\neigenvalue =', eigenvalue)
```
